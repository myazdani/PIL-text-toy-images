```python
import PIL
from PIL import ImageFont
from PIL import Image
from PIL import ImageDraw
import string
```

```python
image = Image.new("RGBA", (150,150), (255,255,255))
fnt =ImageFont.truetype("/System/Library/Fonts/HelveLTMM",size=80)
```

```python
draw = ImageDraw.Draw(image)
```

```python
draw.text((10, 10), "hello", (0,0,0), font = fnt)
```

```python
image.save("hello.png")
```

```python
cats = string.lowercase[:20]
```

```python
cats[0]
```

```python
img_ids = []
for cat in cats:
    img_ids.extend([cat+str(i) for i in range(30)])
```

```python
img_ids
```

```python
draw.text((10, 10), img_ids[0], (0,0,0), font = fnt)
image.save("hello.jpg")
```

```python
for img_id in img_ids:
    image = Image.new("RGBA", (150,150), (255,255,255))
    fnt =ImageFont.truetype("/System/Library/Fonts/HelveLTMM",size=80)  
    draw = ImageDraw.Draw(image)
    draw.text((10, 10), img_id, (0,0,0), font = fnt)
    image.save(img_id + ".png")
```
